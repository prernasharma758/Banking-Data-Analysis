```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd

df = pd.read_csv("Churn_Modelling.csv")
df.head()
```

```python
df = df.drop(columns=["RowNumber","CustomerId","Surname"], errors='ignore')
df.isnull().sum()
df.duplicated().sum()
```

```python
print(df.columns.tolist()),
df.isnull().sum()
df.duplicated().sum()
```

```python
df["Exited"].value_counts()
```

```python
sns.countplot(x="Exited", data=df)
plt.show()
```

```python
df.groupby("Gender")["Exited"].sum()
```

```python
sns.countplot(x="Gender", hue="Exited", data=df)
plt.show()
```

```python
df.groupby("Geography")["Exited"].sum()
```

```python
sns.countplot(x="Geography", hue="Exited", data=df)
plt.show()
```

```python
df["Age"].describe()
```

```python
sns.histplot(df["Age"])
plt.show()
```

```python
df.groupby("IsActiveMember")["Exited"].sum()
```

```python
sns.countplot(x="IsActiveMember", hue="Exited", data=df)
plt.show()
```

- Inactive customers churn more
- Older customers leave more
- Some countries have higher churn
- Gender affects churn

**Conclusion**

This project analyzes bank customers and shows that factors like age, geography, and activity affect customer churn.
